```python
data2 = pd.DataFrame(data)
```

```python
# (b) Now Remove the some of units from Mileage, Engine, Power, and New_Price
```

```python
data2['Mileage'] = data2['Mileage'].str.extract('(\d+\.\d+)').astype(float)

data2['Engine'] = data2['Engine'].str.replace(' CC', '').astype(float)

data2['Power'] = data2['Power'].str.extract('(\d+\.\d+)').astype(float)

data2['New_Price'] = data2['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
# (a)Now Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them.
#Replace the NA values of Mileage, Engine and seats with median
#Replace the NA values of power with Mode.As there are many null values incase of Newprice, it is better to drop a column because imputing any values can result in outliers in the data and lead to wrong analysis
```

```python
data2['Mileage'].fillna(data2['Mileage'].median(),inplace=True)
data2['Engine'].fillna(data2['Engine'].median(),inplace=True)
data2['Power'].fillna(data2['Power'].mode()[0],inplace=True)

data2['Seats'].fillna(data2['Seats'].median(),inplace=True)

data2.drop('New_Price', axis=1, inplace=True)#axis=1 indicates to drop a column
```

```python
output_csv_file = ('/content/clean.csv')
data2.to_csv(output_csv_file, index=False)
```
